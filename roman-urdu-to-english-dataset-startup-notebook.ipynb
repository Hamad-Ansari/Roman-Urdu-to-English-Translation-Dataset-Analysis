## **üöÄ Let's Dive In!**

Prepare for a **colorful, emoji-rich, and fully visual exploration** of the dataset!  

```python
# ==============================================================
# üåü Roman Urdu Sentences Dataset Analysis Notebook üåü
# ==============================================================
# Fully visual, interactive analysis with emojis and beautiful plots

# 1Ô∏è‚É£ Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from wordcloud import WordCloud
from IPython.display import display, Markdown

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['axes.titlesize'] = 16

# 2Ô∏è‚É£ Load Dataset
df = pd.read_csv("/kaggle/input/roman-urdu-sentences-urdu-and-english-translations/Roman_Urdu_Parallel_Translation.csv")
display(Markdown("## üóÇ Dataset Preview"))
display(df.head())

# 3Ô∏è‚É£ Dataset Info & Missing Values
display(Markdown("## ‚ÑπÔ∏è Dataset Info"))
print(df.info())
display(Markdown("## ‚ùå Missing Values Check"))
print(df.isnull().sum())

# 4Ô∏è‚É£ Number of Sentences per Category
display(Markdown("## üìä Sentences per Category"))
category_counts = df['category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title("Sentences Distribution Across Categories")
plt.xlabel("Category")
plt.ylabel("Number of Sentences")
plt.show()

# 5Ô∏è‚É£ Sentence Length Analysis
display(Markdown("## ‚úèÔ∏è Sentence Length Analysis"))

# Add length columns
df['roman_len'] = df['roman_urdu'].apply(lambda x: len(str(x).split()))
df['urdu_len'] = df['urdu_script'].apply(lambda x: len(str(x).split()))
df['english_len'] = df['translated_by_llm'].apply(lambda x: len(str(x).split()))

# Plot histogram per column individually to avoid color mismatch
for col, color in zip(['roman_len','urdu_len','english_len'], ['#2ca02c','#ff7f0e','#1f77b4']):
    plt.hist(df[col], bins=15, color=color, alpha=0.7)
    plt.title(f"Sentence Length Distribution - {col}")
    plt.xlabel("Words per Sentence")
    plt.ylabel("Frequency")
    plt.show()

# 6Ô∏è‚É£ Top Words in Roman Urdu
display(Markdown("## üî§ Most Frequent Words in Roman Urdu"))
all_words = ' '.join(df['roman_urdu'].astype(str)).split()
top_words = Counter(all_words).most_common(15)
for word, count in top_words:
    print(f"üí¨ {word}: {count}")

# 7Ô∏è‚É£ Top Words in English Translations
display(Markdown("## üî§ Most Frequent Words in English"))
all_eng_words = ' '.join(df['translated_by_llm'].astype(str)).split()
top_eng_words = Counter(all_eng_words).most_common(15)
for word, count in top_eng_words:
    print(f"üó®Ô∏è {word}: {count}")

# 8Ô∏è‚É£ Category Distribution Pie Chart
display(Markdown("## ü•ß Category Distribution Pie Chart"))
df['category'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90, title="Category Distribution")
plt.ylabel("")
plt.show()

# 9Ô∏è‚É£ Sentence Length Comparison Boxplot
display(Markdown("## üìè Sentence Length Boxplot"))
sns.boxplot(data=df[['roman_len','urdu_len','english_len']], palette='coolwarm')
plt.title("Sentence Length Comparison Across Languages")
plt.show()

# üîü Word Clouds
display(Markdown("## ‚òÅÔ∏è Word Clouds"))

# Roman Urdu Word Cloud
display(Markdown("### üìù Roman Urdu Word Cloud"))
wc_roman = WordCloud(width=600, height=400, background_color='white', colormap='viridis').generate(' '.join(df['roman_urdu'].astype(str)))
plt.imshow(wc_roman, interpolation='bilinear')
plt.axis('off')
plt.show()

# Urdu Word Cloud
display(Markdown("### üñã Urdu Script Word Cloud"))
wc_urdu = WordCloud(width=600, height=400, background_color='white', colormap='plasma').generate(' '.join(df['urdu_script'].astype(str)))
plt.imshow(wc_urdu, interpolation='bilinear')
plt.axis('off')
plt.show()

# English Word Cloud
display(Markdown("### ‚úçÔ∏è English Translation Word Cloud"))
wc_eng = WordCloud(width=600, height=400, background_color='white', colormap='magma').generate(' '.join(df['translated_by_llm'].astype(str)))
plt.imshow(wc_eng, interpolation='bilinear')
plt.axis('off')
plt.show()

# 1Ô∏è‚É£1Ô∏è‚É£ Random Examples
display(Markdown("## üé≤ Random Examples from Dataset"))
display(df.sample(5))

# 1Ô∏è‚É£2Ô∏è‚É£ Notes for Kaggle Users
display(Markdown("""
## üß© How to Use This Dataset
- Can be used for **Roman Urdu to English translation tasks**  
- Useful for **NLP research, machine translation, and text preprocessing experiments**  
- Ideal for building **Roman Urdu spell checkers, transliteration, or chatbots**  

### üìå Columns Description
| Column | Description |
|--------|-------------|
| category | Category of sentence (Health, Social_Media, etc.) |
| roman_urdu | Sentence in Roman Urdu |
| urdu_script | Sentence in Urdu script |
| translated_by_llm | English translation generated by LLM |
| translated_by_google | English translation from Google Translate (for comparison) |
| roman_len | Number of words in Roman Urdu sentence |
| urdu_len | Number of words in Urdu script sentence |
| english_len | Number of words in English translation |
"""))

display(Markdown("‚úÖ Notebook Complete! Enjoy exploring the dataset üåü"))
```
